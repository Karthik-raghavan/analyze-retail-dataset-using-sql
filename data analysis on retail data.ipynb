```python
# Required Libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sqlalchemy
```

```python
engine = sqlalchemy.create_engine("mysql+pymysql://root:admin@localhost/retail")
```

```python
customers = pd.read_csv("retail-data-datasets/customers.csv")
geolocation = pd.read_csv("retail-data-datasets/geolocation.csv")
orders = pd.read_csv("retail-data-datasets/orders.csv")
order_items = pd.read_csv("retail-data-datasets/order_items.csv")
order_reviews = pd.read_csv("retail-data-datasets/order_reviews.csv", encoding="latin1")
payments = pd.read_csv("retail-data-datasets/payments.csv")
products = pd.read_csv("retail-data-datasets/products.csv")
sellers = pd.read_csv("retail-data-datasets/sellers.csv")
```

```python
customers.to_sql("customers",engine,index=False)
geolocation.to_sql("geolocation",engine,index=False)
orders.to_sql("orders",engine,index=False)
order_items.to_sql("order_items",engine,index=False)
order_reviews.to_sql("order_reviews",engine,index=False)
payments.to_sql("payments",engine,index=False)
products.to_sql("products",engine,index=False)
sellers.to_sql("sellers",engine,index=False)
```

```python
information_schema = "SELECT * FROM INFORMATION_SCHEMA.tables WHERE TABLE_SCHEMA = 'retail'"
pd.read_sql_query(information_schema,engine)
```
